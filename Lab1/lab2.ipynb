```python
class Environment:
  def __init__(self):
# Create a 3x3 grid, where 'b', 'e', and 'f' are dirty
    self.grid = ['enemy near', 'enemy shoot', 'enemy far',
                'health low', 'enemy near', 'health low',
                'enemy far', 'enemy shoot', 'enemy near']
  def get_percept(self, position):
    # Return the state of the current position
    return self.grid[position]
  def clean_room(self, position, action):
    # Clean the room at the given position
    self.grid[position] = action
  def display_grid(self, agent_position):
    # Display the current state of the grid in a 3x3 format
    print("\nCurrent Grid State:")
    grid_with_agent = self.grid[:] # Copy the grid
    grid_with_agent[agent_position] = "游놓" # Place the agent at the current position
    for i in range(0, 9, 3):
      print(" | ".join(grid_with_agent[i:i + 3]))
    print() # Extra line for spacing

class SimpleReflexAgent:
  def __init__(self):
    self.position = 0 # Start at 'a' (position 0 in the grid)
    self.action = {
                    "enemy near": "take cover",
                    "enemy far": "shoot",
                    "health low": "heal",
                    "enemy shoot":"take cover"
                  }
  def act(self, percept, grid):
    grid[self.position] = self.action[percept]
    return self.action[percept]
  def move(self):
    # Move to the next position in the grid
    if self.position < 8:
      self.position += 1
    return self.position

def run_agent(agent, environment, steps):
  for step in range(steps):
    percept = environment.get_percept(agent.position)
    action = agent.act(percept, environment.grid)
    print(f"Step {step + 1}: Position {agent.position} -> Percept- {percept}, Action - {action}")
    environment.display_grid(agent.position) # Display the grid state with agent position
    environment.clean_room(agent.position,agent.action[percept])
    agent.move()
# Create instances of agent and environment
agent = SimpleReflexAgent()
environment = Environment()
# Run the agent in the environment for 9 steps (to cover the 3x3 grid)
run_agent(agent, environment, 9)
```

```python
import random
class Environment:
  def __init__(self):
    self.grid=[random.choice(['safe', 'vulnerable']) for _ in range(9)]
    print(self.grid)
  def get_percept(self, position):
    # Return the state of the current position
    return self.grid[position]
  def display_grid(self):
    print(self.grid)
    for i in range (9):
      if(self.grid[i]!="safe"):
        print("still unsafe")
        return
    print("Completely Safe")

class SimpleReflexAgent:
  def __init__(self):
    self.patching = []

  def system_scan(self, grid):
    for position in range (9):
      if(grid[position]=="vulnerable"):
        print(f'warning at position {position}')
        self.patching.append(position)
      else:
        print(f'success at position {position}')

  def patching_vulnerabilities(self, grid):
    for i in self.patching:
      print(f'patching position {i}')
      grid[i]="safe"

def run_agent(agent, environment):
  agent.system_scan(environment.grid)
  agent.patching_vulnerabilities(environment.grid)
  environment.display_grid()

agent = SimpleReflexAgent()
environment = Environment()
run_agent(agent, environment)
```

```python
import random
class Environment:
  def __init__(self):
    self.grid=[random.choice(['Underloaded','Balanced','Overloaded']) for _ in range(5)]
    print(self.grid)
  def get_percept(self, position):
    return self.grid[position]
  def display_grid(self):
    print(self.grid)
    for i in range (5):
      if(self.grid[i]=="Overloaded"):
        print("still Overloaded, Not enough servers to distribute work")
        return
    print("Workload distributed Successfully")

class LoadBalancerAgent:
  def __init__(self):
    pass

  def scanning_and_balancing(self, grid):
    for i in range (5):
      if(grid[i]!="Overloaded"):
        print(f'Server {i} is fine')
        continue
      print(f'Server {i} is Overloaded')
      for j in range (5):
        if(grid[j]=="Underloaded"):
          grid[j]="Balanced"
          grid[i]="Balanced"
          print(f'Server {i} is balanced with Server {j}')
          break

def run_agent(agent, environment):
  agent.scanning_and_balancing(environment.grid)
  environment.display_grid()

agent = LoadBalancerAgent()
environment = Environment()
run_agent(agent, environment)
```

```python
import random
class Environment:
  def __init__(self):
    self.grid=[random.choice(['safe', 'low risk vulnerable', 'high risk vulnerable']) for _ in range(9)]
    self.purchased=bool(random.choice([True,False]))
    self.id="Hello"
    self.pwd="Hi"
    print(self.grid)

  def get_percept(self, position):
    # Return the state of the current position
    return self.grid[position]
  def display_grid(self):
    print(self.grid)
    for i in range (9):
      if(self.grid[i]!="safe"):
        print("still unsafe as you do not have utility for high risk")
        return
    print("Completely Safe")

class UtilityBasedAgent:
  def __init__(self):
    self.patching = []

  def system_scan(self, grid):
    for position in range (9):
      if(grid[position]=="high risk vulnerable" or grid[position]=="low risk vulnerable"):
        print(f'warning at position {position}')
        self.patching.append(position)
      else:
        print(f'success at position {position}')

  def patching_vulnerabilities(self, grid, purchased, id, pwd):
    flag=True
    if not purchased:
      print("Utility not purchased")
      flag=False
    else:
      idIn = input("Enter ID: ")
      pwdIn = input("Enter Password: ")
      if pwdIn==pwd and idIn==id:
        print("ID and Password Recognized")
      else:
        print("Invalid ID or Password")
        flag=False

    if not flag:
      for i in self.patching:
        if(grid[i]=="low risk vulnerable"):
          print(f'patching position {i}')
          grid[i]="safe"
    else:
      for i in self.patching:
        if(grid[i]!="safe"):
          print(f'patching position {i}')
          grid[i]="safe"


def run_agent(agent, environment):
  agent.system_scan(environment.grid)
  agent.patching_vulnerabilities(environment.grid, environment.purchased, environment.id, environment.pwd)
  environment.display_grid()

agent = UtilityBasedAgent()
environment = Environment()
run_agent(agent, environment)
```

```python
import random

class Environment:
    def __init__(self, rooms, patients, medicine_storage, nurse_stations, corridors, schedules):
        self.rooms = rooms
        self.patients = patients
        self.medicine_storage = medicine_storage
        self.nurse_stations = nurse_stations
        self.corridors = corridors
        self.locations = rooms+medicine_storage+nurse_stations+corridors
        self.staff = bool(random.choice([True,False]))
        self.schedules=schedules

class GoalBasedAgent:
    def __init__(self, environment):
        self.environment = environment
        self.current_location = environment.corridors[0]
        self.carrying_medicine = None

    def move_to(self, location):
      if location in self.environment.locations:
        print(f"Moving from {self.current_location} to {location}.")
        self.current_location = location
        return True
      else:
        print("Invalid destination, Not in my environment")
        return False

    def pick_up_medicine(self, medicine):
        if self.current_location != self.environment.medicine_storage[0]:
          if(not self.move_to(self.environment.medicine_storage[0])):
            return False
        self.carrying_medicine = medicine
        print(f"Picked up medicine: {medicine}.")
        return True

    def scan_patient_id(self, patient_id):
        print(f"Scanning patient ID: {patient_id}.")
        if patient_id in self.environment.patients:
          print("Patient Identified")
          return True
        else:
          self.alert_staff("UnIdentified Patient, Take action accordingly")
          return False

    def deliver_medicine(self, room, patient_id, medicine):
        if not self.check_med_time(patient_id):
          return False
        if self.current_location == room and self.carrying_medicine:
            print("Comming to deliver medicine")
        else:
            if not self.pick_up_medicine(medicine):
              return False
            if not self.move_to(room):
              return False
        if self.scan_patient_id(patient_id):
                print(f"Delivered {self.carrying_medicine} to patient {patient_id} in {room}.")
                self.carrying_medicine = None
        else:
                print("Patient ID mismatch. Cannot deliver medicine.")


    def alert_staff(self, message):
        print(f"Alerting staff: {message}")

    def check_staff(self):
      if not self.environment.staff:
        self.alert_staff("Lack of Staff")

    def check_med_time(self, patient_id):
      current_time=random.choice(['morning', 'evening', 'night'])
      if current_time not in self.environment.schedules[patient_id]:
        self.alert_staff("It is "+current_time+". Giving medicine to patient "+patient_id)
        return True
      else:
        return False

def run_agent(agent, environment):
  agent.deliver_medicine("101", "1001", "Painkiller")

schedules = {
    "1001": ["morning, night"],
    "1002": ["evening"]
}

environment = Environment(
    rooms=["101", "102"],
    patients=["1001", "1002"],
    medicine_storage=["storage_room"],
    nurse_stations=["nurse_station"], corridors=["Corridor1", "Corridor2"],
    schedules=schedules
)

agent = GoalBasedAgent(environment)

run_agent(agent, environment)
```
